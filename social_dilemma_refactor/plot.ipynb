### Reward comparison

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Specify the folder containing the CSV files
folder_path = './results/clean_up_ppo_collective_reward'

# Loop through all files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv') and "reward" in file_name:  # Check if the file is a CSV
        file_path = os.path.join(folder_path, file_name)

        # Read the CSV file
        data = pd.read_csv(file_path, header=None)

        # Use row indices for x-axis and the single column for y-axis
        plt.plot(data.index, data[0], label=file_name)

# Add labels, legend, and title
plt.xlabel('Iteration')
plt.ylabel('Reward')
plt.title(folder_path.split("/")[-1])
plt.legend()
plt.show()
```
