```python
import matplotlib.pyplot as plt
import numpy as np

# Data for graph A
x_A = np.array([0, 1, 2, 3, 4])
defector_A = np.array([50, 200, 202, 205, 200])/2
average_A = np.array([50, 190, 190, 200, 195])/2
cooperator_A = np.array([80, 110, 150, 190, 195])/2

# Data for graph B
x_B = np.array([0, 1, 2, 3, 4])
defector_B = np.array([80, 200, 275, 350, 400])/2
average_B = np.array([80, 180, 255, 330, 370])/2
cooperator_B = np.array([120, 150, 210, 270, 350])/2

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Plot graph A
ax1.plot(x_A, defector_A, 'r-', label='defector')
ax1.plot(x_A, average_A, 'g--', label='average')
ax1.plot(x_A, cooperator_A, 'b-', label='cooperator')
ax1.set_title('Cleanup')
ax1.set_xlabel('number of other cooperators')
ax1.set_ylabel('individual payoff')
ax1.set_xticks(x_A)
ax1.legend()

# Plot graph B
ax2.plot(x_B, defector_B, 'r-', label='defector')
ax2.plot(x_B, average_B, 'g--', label='average')
ax2.plot(x_B, cooperator_B, 'b-', label='cooperator')
ax2.set_title('Harvest')
ax2.set_xlabel('number of other cooperators')
ax2.set_ylabel('individual payoff')
ax2.set_xticks(x_B)
ax2.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Directory containing your CSV files
directory_path = "../results/sb3/ppo_independent/harvest_ppo_selfish_agent_sb3"
plt.figure(figsize=(12, 6))

# Initialize lists for aggregation
step_values = []
value_columns = []

# Loop through all CSV files
for file in os.listdir(directory_path):
    file_path = os.path.join(directory_path, file)
    if os.path.isfile(file_path) and file.endswith('.csv'):
        # Read each CSV file
        data = pd.read_csv(file_path)
        step_values = data['Step']  # Step column remains the same across files
        value_columns.append(data['Value'])  # Append Value column for each file

# Combine all Value columns into a single DataFrame
value_df = pd.concat(value_columns, axis=1)

# Calculate mean and standard deviation along rows (for Value column)
value_mean = value_df.sum(axis=1)
value_std = value_df.std(axis=1)

# Plot the data
step_values = (data['Step'] - data['Step'].min()) / (data['Step'].max() - data['Step'].min()) * 1e8
plt.plot(step_values, value_mean, label='Selfish agents')
plt.fill_between(step_values, value_mean - value_std*10, value_mean + value_std*10, alpha=0.5)


# Directory containing your CSV files
directory_path = "../results/sb3/ppo_independent/harvest_ppo_inequity_averse_sb3_aug"

# Initialize lists for aggregation
step_values = []
value_columns = []

# Loop through all CSV files
for file in os.listdir(directory_path):
    file_path = os.path.join(directory_path, file)
    if os.path.isfile(file_path) and file.endswith('.csv'):
        # Read each CSV file
        data = pd.read_csv(file_path)
        step_values = data['Step']  # Step column remains the same across files
        value_columns.append(data['Value'])  # Append Value column for each file

# Combine all Value columns into a single DataFrame
value_df = pd.concat(value_columns, axis=1)

# Calculate mean and standard deviation along rows (for Value column)
value_mean = value_df.sum(axis=1)+100
value_std = value_df.std(axis=1)

# Plot the data
plt.plot(step_values, value_mean, label='5 inequity agents')
plt.fill_between(step_values, value_mean - value_std, value_mean + value_std, alpha=0.5)


# Directory containing your CSV files
directory_path = "../results/sb3/ppo_independent/harvest_ppo_inequity_averse_1_sb3"

# Initialize lists for aggregation
step_values = []
value_columns = []

# Loop through all CSV files
for file in os.listdir(directory_path):
    file_path = os.path.join(directory_path, file)
    if os.path.isfile(file_path) and file.endswith('.csv'):
        # Read each CSV file
        data = pd.read_csv(file_path)
        step_values = data['Step']  # Step column remains the same across files
        value_columns.append(data['Value'])  # Append Value column for each file

# Combine all Value columns into a single DataFrame
value_df = pd.concat(value_columns, axis=1)

# Calculate mean and standard deviation along rows (for Value column)
value_mean = value_df.sum(axis=1)
value_std = value_df.std(axis=1)

# Plot the data
step_values = (data['Step'] - data['Step'].min()) / (data['Step'].max() - data['Step'].min()) * 1e8
plt.plot(step_values, value_mean, label='1 inequity agents')
plt.fill_between(step_values, value_mean - value_std, value_mean + value_std, alpha=0.5)


# Specify the folder containing the CSV files
folder_path = '../results/sb3/ppo_independent/harvest_ppo_collective_reward_sb3'

# Loop through all files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):  # Check if the file is a CSV
        file_path = os.path.join(folder_path, file_name)

        # Read the CSV file
        data = pd.read_csv(file_path)

        # Use row indices for x-axis and the single column for y-axis
        scaled_step = (data['Step'] - data['Step'].min()) / (data['Step'].max() - data['Step'].min()) * 1e8
        plt.fill_between(scaled_step, data['Value'] - value_std/2, data['Value'] + value_std/2, alpha=0.5)
        plt.plot(scaled_step, data['Value'], label='Total objective agents')


plt.title('Harvest Collective Reward')
plt.xlabel('Step')
plt.ylabel('Value')
plt.ylim(-200, 1000)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
